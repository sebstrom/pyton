### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Objektorienterad
x = np.linspace(-4, 4)
y = x**2

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, label = "X i kvadrat")
ax.set_title("Bra titel")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.legend()
```

```python
# pyplot
x = np.linspace(-4, 4)
y = x**2

plt.figure(figsize=(5, 3))
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Fin graf")
plt.plot(x, y, label = "X i kvadrat")
plt.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

fig, ax = plt.subplot_mosaic([["left", "right"], ["left", "right"]], layout= "constrained")
ax["left"].scatter(random_data_x ,random_data_y)
ax["left"].plot(x,y, color = "red")
ax["left"].set_title("Scatter plot")
# Data for bar plot

fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
colors = ["purple", "red", "orange", "yellow", "green"]
ax["right"].set_title("Bar plot")
ax["right"].bar(names, values, color = colors)

```
